```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
image=cv2.imread("Sasuke.jpg")
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
pixels=image.reshape((-1,3))
pixels = np.float32(pixels)
```

```python
print(pixels)
```

```python
when_stop=(cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,100,0.2)
```

```python
error, labels, centers = cv2.kmeans(data=pixels, K=5,bestLabels=None,criteria=when_stop,attempts=10,flags=cv2.KMEANS_RANDOM_CENTERS)
```

```python
centers = np.uint8(centers)
```

```python
compressed = centers[labels.flatten()]
```

```python
compressed = compressed.reshape(image.shape)
```

```python
cv2.imwrite("compressed.jpg",cv2.cvtColor(compressed,cv2.COLOR_RGB2BGR))
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(image)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Compressed(k=32)")
plt.imshow(compressed)
plt.axis('off')

plt.show()
```
